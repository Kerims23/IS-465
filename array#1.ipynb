#### Create an array of 10 zeros 

```python
array=np.zeros(10)
print("Creating an array of 10 zeros:")
print(array)
```

#### Create an array of 10 ones

```python
array=np.ones(10)
print("An array of 10 ones: ")
print(array)
```

#### Create an array of 10 tens

```python
array=np.ones(10)*10
print("An array of 10 tens: ")
print(array)
```

#### Create an array of the integers from 10 to 100

```python
a=np.arange(10,101)
print(a)
```

#### Create an array of all the odd integers from 10 to 30

```python
a = np.array([i for i in range(10, 31) if i%2 != 0])
print(a)
```
